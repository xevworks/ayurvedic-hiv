# 4. Process to search only HIV

```python
import pandas as pd
```

```python
file_01 = r'E:\research\ayurvedic-hiv\data\cleaned\08_ayurvedic-knapsack-bindingdb-exploded-cleaned.csv'
```

```python
df = pd.read_csv(file_01)
df
```

```python
df_hiv = df[df['species'].str.contains(r'Human immunodeficiency virus type 1|HIV-1', na=False)]
df_hiv
```

```python
len(df_hiv['C_ID'].unique())
```

```python
df_hiv.drop_duplicates(inplace=True)
df_hiv
```

```python
df_hiv.isna().sum()
```

```python
df_hiv.to_csv(r'E:\research\ayurvedic-hiv\data\processed\09_ayurvedic-knapsack-bindingdb-hiv.csv', index=False)
```

keywords = ["protease", "transcriptase", "integrase", "glycoprotein"]

```python
keywords = ["protease", "transcriptase", "integrase", "glycoprotein"]
pattern = "|".join(keywords)

df_hiv_target = df_hiv[df_hiv['target_name'].str.contains(pattern, case=False, na=False)]
df_hiv_target
```
