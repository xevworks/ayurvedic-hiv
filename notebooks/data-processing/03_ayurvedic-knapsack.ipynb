# 2. Processing Ayurvedic-KNaPSAcK data

```python
import pandas as pd
```

```python
file_01 = r'E:\research\ayurvedic-hiv\data\original\03_ayurvedic-knapsack.csv'
```

```python
df_ayur_knap = pd.read_csv(file_01, index_col=0)
df_ayur_knap
```

```python
df_ayur_knap.isna().sum()
```

## 2.1 Cleaning data of Ayurvedic-KNApSAcK

```python
isna_not_allowed_cols = ['Formula', 'CAS RN', 'InChIKey', 'InChICode', 'SMILES']
df_ayur_knap_clean = df_ayur_knap.dropna(subset=isna_not_allowed_cols)
df_ayur_knap_clean.isna().sum()
```

### 2.1.1 Give code for Ayurvedic-KNApSAcK relation

```python
# Give a unique FSM_Code to each row -> FSM means Formula-Secondary Metabolite
df_ayur_knap_clean['FSM_Code'] = ['AyKn' + str(i).zfill(6) for i in range(1, len(df_ayur_knap_clean) + 1)]

cols = list(df_ayur_knap_clean.columns)
cols.insert(1, cols.pop(cols.index('FSM_Code')))
df_ayur_knap_clean = df_ayur_knap_clean[cols]
df_ayur_knap_clean
```

```python
len(df_ayur_knap_clean['C_ID'].unique())
```

```python
df_ayur_knap_clean.isna().sum()
```

```python
# Save the cleaned DataFrame to a new CSV file
df_ayur_knap_clean.to_csv(r'E:\research\ayurvedic-hiv\data\cleaned\04_ayurvedic-knapsack-clean.csv', index=False)
```

## 2.2 Grouping Ayurvedic into Secondary Metabolite (KNApSAcK [C_ID])

```python
df_ayur_knap_clean_grouped = df_ayur_knap_clean.groupby('C_ID').agg({
    'Code': lambda x: ', '.join(x.dropna().astype(str).unique()),
    'FSM_Code': lambda x: ', '.join(x.dropna().astype(str).unique()),
    'Name': lambda x: ', '.join(x.dropna().astype(str).unique()),
    'Common Name': lambda x: ', '.join(x.dropna().astype(str).unique()),
    'Scientific Name': lambda x: ', '.join(x.dropna().astype(str).unique()),
    'Quantity': lambda x: ', '.join(x.dropna().astype(str).unique()),
    'Non plant ingredient': lambda x: ', '.join(x.dropna().astype(str).unique()),
    'Action': lambda x: ', '.join(x.dropna().astype(str).unique()),
    'Application': lambda x: ', '.join(x.dropna().astype(str).unique()),
    'Comments': lambda x: ', '.join(x.dropna().astype(str).unique()),
    'Bacteria': lambda x: ', '.join(map(str, sorted(x.dropna().unique()))),
    'Metabolite Name': 'first',  # Unique to each C_ID
    'Formula': 'first',
    'Mw': 'first',
    'CAS RN': 'first',
    'InChIKey': 'first',
    'InChICode': 'first',
    'SMILES': 'first'
}).reset_index()
df_ayur_knap_clean_grouped
```
