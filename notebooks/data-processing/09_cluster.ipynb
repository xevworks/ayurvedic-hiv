```python
cek["cluster"] = cek["cluster"].str.split(",")

# 2) Explode: setiap elemen di cluster_list menjadi baris sendiri
cek_exploded = cek.explode("cluster")

# 3) Hapus spasi berlebih di depan/akhir, sehingga " 4" â†’ "4"
cek_exploded["cluster"] = cek_exploded["cluster"].str.strip()
```

```python
# 4) Ubah format cluster_list jadi integer (jika memang angka)
cek_exploded["cluster"] = cek_exploded["cluster"].astype(int)

# 5) Jika ingin groupby per nilai cluster, misalnya daftar semua C_ID, Code, dan FSM_Code per cluster
cek_grouped = (
    cek_exploded
    .groupby("cluster")  # group berdasarkan tiap cluster (2, 3, 4, 5)
    .agg({"C_ID": list, "Code": list, "FSM_Code": list})  # gabungkan semua C_ID, Code, dan FSM_Code ke dalam satu list
    .reset_index()
)

cek_grouped
```
