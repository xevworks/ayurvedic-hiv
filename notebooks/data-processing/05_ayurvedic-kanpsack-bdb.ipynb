```python
file_01 = r'E:\research\ayurvedic-hiv\data\processed\06_ayurvedic-knapsack-bindingdb.csv'
```

```python
df_ayur_knap_bdb = pd.read_csv(file_01, index_col=0)
df_ayur_knap_bdb 
```

## 3.1 Cleaning Ayurvedic-KNApSAcK-BDB data from NaN

```python
cols_to_check = ['target_name', 'species', 'bdb_id', 'ligand_smiles', 'pdb_link']

df_ayur_knap_bdb = df_ayur_knap_bdb.dropna(subset=cols_to_check)
df_ayur_knap_bdb.isna().sum()
```

### 3.1.1 Cleaning where column value has 'No Similarity Matches'

```python
df_ayur_knap_bdb = df_ayur_knap_bdb[~df_ayur_knap_bdb[cols_to_check].isin(['No Similarity Matches']).any(axis=1)]
df_ayur_knap_bdb
```

```python
df_ayur_knap_bdb.to_csv(r'E:\research\ayurvedic-hiv\data\cleaned\07_ayurvedic-knapsack-bindingdb-cleaned.csv', index=False)
```
